#Start Model Train


###Train-test Spilit

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(clean_data[['Age','Gender','self_employed','family_history','work_interfere','remote_work','tech_company','benefits','care_options','wellness_program','seek_help','leave','mental_health_consequence','phys_health_consequence','coworkers','supervisor','mental_vs_physical','obs_consequence']],clean_data['treatment'],test_size=0.20)
```

```python
len(x_test)
```

```python
len(y_train)
```

```python
len(y_test)
```

#LogisticRegression


```python
from sklearn.linear_model import LogisticRegression
```

```python
lr=LogisticRegression()
```

```python
lr.fit(x_train,y_train)
```

```python
y_pred_lr = lr.predict(x_test)
from sklearn.metrics import classification_report, confusion_matrix
print(confusion_matrix(y_test, y_pred_lr))
print(classification_report(y_test,y_pred_lr))
```

#Decision Tree

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
dt=DecisionTreeClassifier()
```

```python
dt.fit(x_train,y_train)
```

```python
y_pred_dt = dt.predict(x_test)
from sklearn.metrics import classification_report, confusion_matrix
print(confusion_matrix(y_test, y_pred_dt))
print(classification_report(y_test,y_pred_dt))
```

#Random Forest

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
rf= RandomForestClassifier(n_estimators=100, random_state=42)
```

```python
rf.fit(x_train,y_train)
```

```python
y_pred_rf = rf.predict(x_test)
from sklearn.metrics import classification_report, confusion_matrix
print(confusion_matrix(y_test, y_pred_rf))
print(classification_report(y_test,y_pred_rf))
```

#AdaBoostClassifier

```python
from sklearn.ensemble import AdaBoostClassifier
```

```python
adb = AdaBoostClassifier(n_estimators=50, random_state=42)
```

```python
adb.fit(x_train,y_train)
```

```python
y_pred_abd = adb.predict(x_test)
from sklearn.metrics import classification_report, confusion_matrix
print(confusion_matrix(y_test, y_pred_abd))
print(classification_report(y_test,y_pred_abd))
```
